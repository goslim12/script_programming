```python
def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:],"")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:],"")
    
    return string
```

```python
#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:],"")
    return string
```

```python
#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:],"")
    return string
```

```python
#구두문자 제거 함수
def punc_delete(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i],"")
        
    return source
```

<li>요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.<ul>
<li>예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---&gt; {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}</li>
</ul>
</li>

```python
#사전형태로 출현빈도 저장 함수
def word_count(words):
    words_dict = {}
    for i in words:
        if words_dict.has_key(i) is False:
            words_dict[i]=1
        else:
            words_dict[i]+=1
    return words_dict
```

```python
#cse_words_count={}
words = word_count(words)
```

```python
for i in words:
    print str(i)+" : "+str(words[i])
```

```python
source_naver = urllib2.urlopen("http://www.naver.com/").read()
source_daum = urllib2.urlopen("http://www.daum.net/").read()
source_nytimes = urllib2.urlopen("http://www.nytimes.com/").read()
```

```python
#스타일시트 제거
source_naver = stylesheet_delete(source_naver)
source_daum = stylesheet_delete(source_daum)
source_nytimes = stylesheet_delete(source_nytimes)
```

```python
#자바스크립트 제거
source_naver = javascript_delete(source_naver)
source_daum = javascript_delete(source_daum)
source_nytimes = javascript_delete(source_nytimes)
```

```python
#HTML 태그제거
source_naver = tag_delete(source_naver)
source_daum = tag_delete(source_daum)
source_nytimes = tag_delete(source_nytimes)
```

```python
#구두문자 제거
source_naver = punc_delete(source_naver)
source_daum = punc_delete(source_daum)
source_nytimes = punc_delete(source_nytimes)
```

```python
#공백을 기준으로 각 단어 분리
words_naver = source_naver.split()
words_daum = source_daum.split()
words_nytimes = source_nytimes.split()
```

```python
#사전형태로 단어 출현빈도 저장
words_naver = word_count(words_naver)
words_daum = word_count(words_daum)
words_nytimes = word_count(words_nytimes)
```
